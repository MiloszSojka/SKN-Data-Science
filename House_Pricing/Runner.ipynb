<h1>Imputation of missing values</h1>
<h3>Wprowadzanie brakujących danych</h3>

Większość algorytmów machine learningowych wymaga danych bez pustych wartości

TODO - msno <- wizualizacje brakujących danych

```python
# Visualising percentage of missing values of the top 10 numeric variables
# total = numerical_features.isnull().sum.sort_values(ascending=False)
# percent = (numerical_features.isnull().sum())/(numerical_features.isnull().count().sort_values(ascending=False))
total_mnd = trainData.select_dtypes(include=[np.number]).isnull().sum().sort_values(ascending=False)
percent_mnd = (trainData.select_dtypes(include=[np.number]).isnull().sum()/trainData.select_dtypes(include=[np.number]).isnull().count()).sort_values(ascending=False)


missing_numerical_data = pd.concat([total_mnd, percent_mnd], axis=1,join='outer', keys=['Missing Count', 'Missing Percentage'])
missing_numerical_data.index.name =' Numeric Feature'
missing_numerical_data.head(10)
```

We observe that 'LotFrontage', 'GarageYrBlt' and 'MasVnrArea' are the only one who have missing values

```python
# Visualising percentage of missing values of the top 10 categorical variables
total_mtd = trainData.select_dtypes(include=[np.object]).isnull().sum().sort_values(ascending=False)
percent_mtd = (trainData.select_dtypes(include=[np.object]).isnull().sum()/trainData.select_dtypes(include=[np.object]).isnull().count()).sort_values(ascending=False)

missing_text_data = pd.concat([total_mtd, percent_mtd], axis=1,join='outer', keys=['Missing Count', 'Missing Percentage'])
missing_text_data.index.name =' Categorical Feature'
missing_text_data.head(10)
```
